##### Interaction Analysis

```python
feature1='# Major Vessels'
nickname1='Number of Major Vessels'
feature2='Max Heart Rate'
nickname2='Max Heart Rate Achieved'

fig,axes,summary_df=info_plots.target_plot_interact(
    df=df,features=[feature1,feature2],feature_names=[nickname1,nickname2],target='Target'
)

axes['value_ax'].set_xticklabels(['0','1','2'])
plt.show()
```
