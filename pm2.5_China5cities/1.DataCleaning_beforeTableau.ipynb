### Beijing

###### Read Data

```python
import pandas as pd
BJ=pd.read_csv('BeijingPM20100101_20151231.csv')  
BJ.head(5)
```

```python
# Drop column
BJ.drop(['PM_Dongsi'], axis=1, inplace=True)
```

```python
# import os
# os.remove('PM2.5.csv')
```

```python
# def label to change season numerical representer to char
def label (row):
    if row['season'] == 1 :
        return 'Spring'
    elif row['season'] == 2 :
        return 'Summer'
    elif row['season'] == 3 :
        return 'Fall'
    else:
        return 'Winter'
```

```python
BJ['Season'] = BJ.apply (lambda row: label(row), axis=1)
```

```python
# Add city column
BJ['City']="Beijing"
```

```python
BJ.iloc[18000:18005]
```

```python
# Combine year month day hour. Fomrat to datetime
BJ['Time'] = BJ.year.map(str)+"/"+BJ.month.map(str)+"/"+BJ.day.map(str)+" "+BJ.hour.map(str)+":00"
#BJ.insert(5, 'Time', (BJ['year'])+(BJ['month'])+(BJ['day'])+(BJ['hour'])+":00"))
BJ['Time'] = pd.to_datetime(BJ['Time'], format='%Y/%m/%d %H:00')
```

```python
print(BJ.dtypes)
```

```python
# Drop columns
BJ.drop(['No','year', 'month','day','hour','season'], axis=1, inplace=True)
```

```python
# rename column to ensure all 5 files will in the end have the same column names
BJ.rename(columns={'PM_US Post':'PM_measurepoint1','PM_Dongsihuan':'PM_measurepoint2','PM_Nongzhanguan':'PM_measurepoint3'}, inplace=True)
BJ.tail(5)
```

```python
# def average to calculate average pm measurement
def average (row):
    if row['PM_measurepoint1']=="Null" and row['PM_measurepoint2']=="Null" and row['PM_measurepoint3']=="Null":
        return 'Null'
    
    elif row['PM_measurepoint1']!="Null" and row['PM_measurepoint2']=="Null" and row['PM_measurepoint3']=="Null":
        return row['PM_measurepoint1']
    elif row['PM_measurepoint2']!="Null" and row['PM_measurepoint1']=="Null" and row['PM_measurepoint3']=="Null":
        return row['PM_measurepoint2']
    elif row['PM_measurepoint3']!="Null" and row['PM_measurepoint1']=="Null" and row['PM_measurepoint2']=="Null":
        return row['PM_measurepoint3']
    
    elif row['PM_measurepoint1']!="Null" and row['PM_measurepoint2']!="Null" and row['PM_measurepoint3']=="Null":
        return (row['PM_measurepoint1']+row['PM_measurepoint2'])/2
    elif row['PM_measurepoint1']!="Null" and row['PM_measurepoint2']=="Null" and row['PM_measurepoint3']!="Null":
        return (row['PM_measurepoint1']+row['PM_measurepoint3'])/2
    elif row['PM_measurepoint2']!="Null" and row['PM_measurepoint3']!="Null" and row['PM_measurepoint1']=="Null":
        return (row['PM_measurepoint2']+row['PM_measurepoint3'])/2
    
    else:
        return (row['PM_measurepoint1']+row['PM_measurepoint2']+row['PM_measurepoint3'])/3
```

```python
df_final['Average'] = df_final.apply (lambda row: average(row), axis=1)
```
